```python
#detect the text of a pic of a second 

def process_video_text_all(path,start_time):
    capture = cv2.VideoCapture(path)
    #print(capture.isOpened())
    fps = capture.get(cv2.CAP_PROP_FPS)
    #print(fps)### 获取 1 秒有多少帧
    total_ = capture.get(cv2.CAP_PROP_FRAME_COUNT)
    #print('total', total_)
    fourcc = cv2.VideoWriter_fourcc('m','p','4','v')
    # width = int(cap.get(cropped.CAP_PROP_FRAME_WIDTH))
    # height = int(cap.get(cropped.CAP_PROP_FRAME_HEIGHT))
    sz = (200, 140)
    outfile_name = path.split('.')[-2]
    outfile_name = outfile_name.split('\\')[-1]
    #print(outfile_name)
    #print('../edicted/'+outfile_name+suffix+'.mp4')
    outfile = cv2.VideoWriter(output_path + outfile_name+'_Discussion'+'.mp4',fourcc, fps, sz)
    outfile_1 = cv2.VideoWriter(output_path + outfile_name+'_Crossword'+'.mp4',fourcc, fps, sz)
    outfile_2 = cv2.VideoWriter(output_path + outfile_name+'_Memory'+'.mp4',fourcc, fps, sz)
    detected = 0 # detect discussion
    detected_1 = 0 #detect crossword
    detected_2 = 0 #detect memory
    
    timer = 0
    min_from = 0
    min_from_1 = 0
    min_from_2 = 0
    print_flag = 0
    print_flag_1 = 0
    while(capture.isOpened()):
          #print('here')## 检测视频打开是否有问题
        timer += 1  ## 开始计算过去了多少帧
        # if timer % 10000 ==0:
        #     print(timer)
            
        ret, frame = capture.read()  ## 逐帧读取视频
        if ret:
            ### detect discussion part
            if (detected == 0) & (timer > start_time*60*fps):
                #print('start_detecting_discussion')
                frame_cut = frame[:300, ::]  #cut the upper part to detect text
                #detect text
                if timer%14 == 0:
                    gray = cv2.cvtColor(frame_cut, cv2.COLOR_BGR2GRAY) 
                    text = pytesseract.image_to_string(gray, lang='eng', config='--psm 6').lower()

                    if (detected == 0 ) & ("team icebreaker activity" in text) : #find discussion part 
                        detected = 1
                        timer_pin = timer # get the time when discussion part start
                        min_from = round(timer_pin/fps/60)
                    #print('min_from', min_from)


            elif detected == 1: # find the discussion part and start to save the vedio
#                 print('capture timer: ', timer)
                if min_from * fps * 60 < timer <  (min_from+7) * fps * 60:
                    #print('comfirm capture',timer)
                    cropped = frame[-190:-50, -200::]  ## 逐帧读取视频
                    outfile.write(cropped)
                elif timer > (min_from+7) * fps * 60: #discussion part finish
                    #print('discussion finished at :' ,(min_from+7) * fps * 60 )
                    outfile.release()
                    detected = 2
#                     print('detected:', detected)
#                     print('detected_1', detected_1)
#                     print('detected_2', detected_2)
            
            elif (detected == 2) & (detected_1 == 0) & (timer > (min_from+7+5.5) * fps * 60): #4.5
                #print('start here')#3 mins after discussion part finish, start to detect crossword    #& (timer > (min_from+2+3) * fps * 60)
                frame_cut_1 = frame[:500, ::]
                if timer % 14 == 0:
                    gray_1 = cv2.cvtColor(frame_cut_1, cv2.COLOR_BGR2GRAY) 
                    text_1 = pytesseract.image_to_string(gray_1, lang='eng', config='--psm 6').lower()

                    if (detected_1 == 0 ) & ("across" in text_1) & ('down' in text_1) : #find crosstab part 

                        detected_1 = 1
                        timer_pin_1 = timer # get the time when crosstab start
                        min_from_1 = round(timer_pin_1/fps/60)



            elif (detected == 2) & (detected_1 == 1):
                if min_from_1 * fps * 60 < timer <  (min_from_1+25) * fps * 60:
                    #print('comfirm capture crossword',timer)
                    cropped = frame[-190:-50, -200::]  ## 逐帧读取视频
                    outfile_1.write(cropped)
                elif timer  >(min_from_1+25) * fps * 60: #crosstab part finish
                    #print('crossword finished', (min_from_1+25) * fps * 60)
                    outfile_1.release()
                    detected_1 = 2                
#                     print('detected:', detected)
#                     print('detected_1', detected_1)
#                     print('detected_2', detected_2)
                    
            elif (detected == 2) & (detected_1 == 2) & (detected_2 == 0) & (timer > (min_from_1+25+1) * fps * 60): #3 mins after discussion & corssword part finish, start to detect memory         
                #print('start')
                frame_cut_2 = frame[:, ::]
                if timer % 14 == 0:
                    gray_2 = cv2.cvtColor(frame_cut_2, cv2.COLOR_BGR2GRAY) 
                    text_2 = pytesseract.image_to_string(gray_2, lang='eng', config='--psm 6').lower()
    #                 plt.figure(figsize=(30,30))
    #                 plt.imshow(frame_cut_2)
    #                 plt.show()
    #                 print(text_2)
                    if (detected_2 == 0 ) & ("algebra" in text_2) : #find memory part 
                        detected_2 = 1
                        
                        # if print_flag == 0:
                        #     print_flag =1
                        #     print('plot of momery 2')
                        #     print('timer', timer)
                        #     print(detected_2)
                        #     print(text_2)
                        #     plt.figure(figsize=(30,30))
                        #     plt.imshow(frame_cut_2)
                        #     plt.show()
                        
                        timer_pin_2 = timer # get the time when crosstab start
                        min_from_2 = round(timer_pin_2/fps/60)
                        

                    
            elif (detected == 2) & (detected_1 == 2) & (detected_2 == 1) :
                # if print_flag_1 == 0:
                #     print_flag_1 = 1
                #     print('here')
                #     print('min_from_2',min_from_2)
                #     print('current',timer)
                if (min_from_2+5) * fps * 60 < total_:
                    if min_from_2 * fps * 60 < timer <  (min_from_2+5) * fps * 60:
                        #print('comfirm capture memory',timer)
                        cropped = frame[-190:-50, -200::]  ## 逐帧读取视频
                        outfile_2.write(cropped)
                    elif timer  > (min_from_2+5) * fps * 60: #crosstab part finish
                        #print('memory finished',min_from_2+5)
                        outfile_2.release()
                        detected_2 = 2
                        break
                else:
                    #print('memory finished',total_)
                    outfile_2.release()
                    detected_2 = 2

        else:
            break
            outfile.release()
            outfile_1.release()
            outfile_2.release()
                

    capture.release()  ## 清内存，好习惯
    cv2.destroyAllWindows()

```
