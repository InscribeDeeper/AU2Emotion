```python
import cv2 
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
min_from = 5 
min_end = 10

capture = cv2.VideoCapture('S20G2A_Webcam_Alpha.mp4')
fps = round(capture.get(5))  ### 获取 1 秒有多少帧
fourcc = cv2.VideoWriter_fourcc('m','p','4','v')
sz = (200, 140)
outfile = cv2.VideoWriter('truncated_S20G2A_Webcam_Alpha.mp4',fourcc, fps, sz)

timer = 0
while(capture.isOpened()):  ## 检测视频打开是否有问题
    timer += 1  ## 开始计算过去了多少帧
    
    if 60 * min_from * fps < timer < 60 * min_end * fps :
        ret, frame = capture.read()  ## 逐帧读取视频
        cropped = frame[-190:-50, -200::]
        outfile.write(cropped) ## outfile 就是创建VideoWriter对象
        
    if timer > 60 * min_end * fps:
        break
        
        
# capture.release()  ## 清内存，好习惯
# outfile.release()
```

```python
plt.figure(figsize=(50,20))
plt.imshow(cropped)
plt.show()
```
